# Point Reflection

You are given a set of points and you need to find the reflection of each point across another specified point.

Consider two points, A(px, py), and B(qx, qy). We define the reflection of point B across point A as a rotation of point B around point A.

Write a function `findPoint` to calculate the reflection of a point B across another point A.

## Function Signature:
```python
def findPoint(px: int, py: int, qx: int, qy: int) -> List[int]:
    pass
```

### Input:
- The function `findPoint` takes four parameters:
  - `px`: an integer representing the x-coordinate of point A.
  - `py`: an integer representing the y-coordinate of point A.
  - `qx`: an integer representing the x-coordinate of point B.
  - `qy`: an integer representing the y-coordinate of point B.

### Output:
- The function should return a list containing two integers representing the coordinates of the reflected point of B across A.

### Constraints:
- -1000 ≤ px, py, qx, qy ≤ 1000

### Example:

#### Input:
```python
px = 1
py = 1
qx = 2
qy = 2
```

#### Output:
```
3 3
```

### Explanation:
- The given point A is (1, 1) and point B is (2, 2).
- The reflection of point B across point A can be calculated as follows:
  - Calculate the difference between qx and px (dx) and qy and py (dy).
  - Add dx and dy to qx and qy respectively to get the coordinates of the reflected point.
- The reflected point coordinates are (3, 3).


````python
def findPoint(px, py, qx, qy):
    # Write your code here
    rx = qx + (qx - px)
    ry = qy + (qy - py)
    return [rx, ry]
````

````python
import matplotlib.pyplot as plt


def plot_point(func):
    def wrapper(*args, **kwargs):
        px, py, qx, qy = args
        rx, ry = func(*args, **kwargs)

        plt.plot(px, py, "ro", label="Point P")
        plt.plot(qx, qy, "bo", label="Point Q")
        plt.plot(rx, ry, "go", label="Point R")

        plt.plot([px, qx], [py, qy], "b-", label="Line PQ")
        plt.plot([qx, rx], [qy, ry], "g-")

        plt.grid(True)
        plt.title("Reflection of Point B across Point A")
        plt.legend()
        plt.axis("equal")

        plt.show()

        return rx, ry

    return wrapper


@plot_point
def findPoint(px, py, qx, qy):
    dx = qx - px
    dy = qy - py

    rx = qx + dx
    ry = qy + dy

    return rx, ry
````

````python
px, py, qx, qy = 0, 0, 1, 1

rx, ry = findPoint(px, py, qx, qy)  # [2, 2]
````
